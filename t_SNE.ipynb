```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("/content/drive/MyDrive/MNIST_DATA.csv")
```

```
df.shape
```

```
df.head()
```

```
l = df["label"]
```

```
d = df.drop(labels = "label",axis=1)
```

```
d.shape
```

```
l.shape
```

```
# data preprocessing  : standardizing the data
from sklearn.preprocessing import StandardScaler
std_data = StandardScaler().fit_transform(d)
print(std_data.shape)
```

# t-SNE using Scikit-learn

```
# TSNE
from sklearn.manifold import TSNE
# picking the top 1.5k data point
data_15k = std_data[0:1500,:]
label_15k = l[0:1500]

model = TSNE(n_components=2,random_state=0)

tsne_data = model.fit_transform(data_15k)

```

```
#creating new dataframe which help us in plotting result data
# appendind labels to 2nd projected data
tsne_data = np.vstack((tsne_data.T,label_15k)).T
# creating new dataframe for plotting labelled points
tsne_df = pd.DataFrame(data=tsne_data,columns=("Dim_1","Dim_2","label"))
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7,7))
sns.FacetGrid(tsne_df,hue = "label",size=6).map(plt.scatter,"Dim_1","Dim_2").add_legend()
plt.show()
```

```
data_5k = std_data[0:5000,:]
label_5k = l[0:5000]

model = TSNE(n_components=2,random_state=0)

tsne_data = model.fit_transform(data_5k)

# appendind labels to 2nd projected data
tsne_data = np.vstack((tsne_data.T,label_5k)).T
# creating new dataframe for plotting labelled points
tsne_df = pd.DataFrame(data=tsne_data,columns=("Dim_1","Dim_2","label"))

plt.figure(figsize=(7,7))
sns.FacetGrid(tsne_df,hue = "label",size=6).map(plt.scatter,"Dim_1","Dim_2").add_legend()
plt.show()
```
